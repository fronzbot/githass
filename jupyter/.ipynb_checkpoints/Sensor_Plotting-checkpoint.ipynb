```python
'''
Setup
'''
import pymysql
import getpass

dbname = 'hassdb'
dbhost = 'localhost'
dbuid  = input("UserID:")
dbpass = getpass.getpass("Password for "+dbuid+":")
```

```python
'''
Sensor to plot and location of database
'''
plot_key  = 'Voltage'

entity_hash = {'Download':['sensor.speedtest_download'],
               'Upload':['sensor.speedtest_upload'],
               'Current':['sensor.power_mon_1_current_4_20'],
               'Voltage':['sensor.power_mon_1_voltage_4_16'],
               'Power': ['sensor.power_mon_1_power_4_8'],
               'Temperature': ['sensor.living_room_temperature', 'sensor.bedroom_temperature'],
               'Humidity': ['sensor.living_room_humidity', 'sensor.pws_relative_humidity']}


entities   = entity_hash[plot_key]
data_type  = plot_key
```

```python
%pylab inline
import datetime
from matplotlib import dates
font = {'weight' : 'normal',
        'size'   : 10}
matplotlib.rc('font', **font)
matplotlib.rc('lines', linewidth=2)
matplotlib.rc({'axes.titlesize' : 'medium'})
conn = pymysql.connect(host=dbhost, user=dbuid, passwd=dbpass, db=dbname)
cursor = conn.cursor()
```

```python
values     = {}
timestamps = {}
dt_list    = {}
for entity in entities:
    cursor.execute("SELECT state, last_changed FROM states WHERE entity_id = '"+entity+"' AND state != 'unknown'")
    values[entity]     = []
    timestamps[entity] = []
    dt_list[entity]    = []
    count = 0;
    for x in cursor.fetchall():
        custom_date = round(dates.date2num(x[1]), 3)
        dt_list[entity].append(x[1])
        timestamps[entity].append(custom_date)
        values[entity].append(float(x[0]))
    
```

```python
figure(num=None, figsize=(12,8), dpi=200) 
plt.hold(True)
plt.grid(b = 'on')
plt.style.use('fivethirtyeight')
for entity in entities:
    plt.plot_date(x=timestamps[entity], y=values[entity], fmt="-", label=entity)
    
plt.legend(loc='lower left')
plt.ylabel(data_type)
plt.xlabel('Time')
```

```python
'''
Plot derivative of data in hourly increments
'''
dx = {}
dy = {}

figure(num=None, figsize=(12,8), dpi=200)  
for entity in entities:
    index_start = 0
    dx[entity] = []
    dy[entity] = []
    for index in range(1,len(values[entity])):
        time_delta = dt_list[entity][index]-dt_list[entity][index_start]
        if time_delta.seconds/3600 >= 1:
            dx[entity].append(timestamps[entity][index])
            dy[entity].append(values[entity][index]-values[entity][index_start])
            index_start = index

         
    plt.plot_date(x=dx[entity], y=dy[entity], fmt="-", label=entity)
    
plt.grid(b = 'on')
plt.style.use('fivethirtyeight')
plt.legend(loc='lower left')
plt.ylabel('Slope of '+data_type+' per Hour')
plt.xlabel('Time')
```
