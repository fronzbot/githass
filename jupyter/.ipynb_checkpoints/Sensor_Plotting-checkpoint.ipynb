```python
'''
Setup
'''
db_path   = '/home/hass/.homeassistant/home-assistant_v2.db'
```

```python
'''
Sensor to plot and location of database
'''
plot_key  = 'Power'
useThresh = 0
HIGH = 700
LOW = 0

entity_hash = {'Download':['sensor.speedtest_download'],
               'Upload':['sensor.speedtest_upload'],
               'Current':['sensor.power_mon_1_current_4_20'],
               'Voltage':['sensor.power_mon_1_voltage_4_16'],
               'Power': ['sensor.power_mon_1_power_4_8'],
               'Temperature': ['sensor.living_room_temperature', 'sensor.bedroom_temperature'],
               'Humidity': ['sensor.living_room_humidity', 'sensor.pws_relative_humidity']}


entities   = entity_hash[plot_key]
data_type  = plot_key
```

```python
%pylab inline
import sqlite3
import datetime
from matplotlib import dates
font = {'weight' : 'normal',
        'size'   : 10}
matplotlib.rc('font', **font)
matplotlib.rc('lines', linewidth=2)
matplotlib.rc({'axes.titlesize' : 'medium'})
conn = sqlite3.connect(db_path)
```

```python
values     = {}
timestamps = {}
dt_list    = {}
thresh     = {}
thresh_time = {}
for entity in entities:
    data = conn.execute("SELECT state, last_changed FROM states WHERE entity_id = '"+entity+"' AND state != 'unknown'")
    values[entity]     = []
    timestamps[entity] = []
    dt_list[entity]    = []
    thresh[entity]     = []
    thresh_time[entity] = []
    count = 0;
    index = 0;
    last = None
    for x in data:
        #custom_date = round(dates.date2num(x[1]), 3)
        dt_list[entity].append(datetime.datetime.strptime(x[1], '%Y-%m-%d %H:%M:%S.%f'))
        timestamps[entity].append(dates.date2num(datetime.datetime.strptime(x[1], '%Y-%m-%d %H:%M:%S.%f')))
        values[entity].append(float(x[0]))
        now   = dt_list[entity][index]
        timedelta = now-dt_list[entity][index-count-1]
        if useThresh is not None:
            if float(x[0]) > useThresh:
                thresh[entity].append(HIGH)
            else:
                thresh[entity].append(LOW)

                
    
```

```python
figure(num=None, figsize=(12,8), dpi=200) 
plt.hold(True)
plt.grid(b = 'on')
plt.style.use('fivethirtyeight')
for entity in entities:
    plt.plot_date(x=timestamps[entity], y=values[entity], fmt="-", label=entity)

plt.plot_date(x=timestamps[entity], y=thresh[entity], fmt="-", label=entity)    
plt.legend(loc='lower left')
plt.ylabel(data_type)
plt.xlabel('Time')
```

```python
'''
Plot derivative of data in hourly increments
'''
dx = {}
dy = {}

figure(num=None, figsize=(12,8), dpi=200)  
for entity in entities:
    index_start = 0
    dx[entity] = []
    dy[entity] = []
    for index in range(1,len(values[entity])):
        time_delta = dt_list[entity][index]-dt_list[entity][index_start]
        if time_delta.seconds/3600 >= 1:
            dx[entity].append(timestamps[entity][index])
            dy[entity].append(values[entity][index]-values[entity][index_start])
            index_start = index

         
    plt.plot_date(x=dx[entity], y=dy[entity], fmt="-", label=entity)
    
plt.grid(b = 'on')
plt.style.use('fivethirtyeight')
plt.legend(loc='lower left')
plt.ylabel('Slope of '+data_type+' per Hour')
plt.xlabel('Time')
```

```python
'''
Plot rolling averge over N minutes
'''
ix = {}
iy = {}

N=2

figure(num=None, figsize=(12,8), dpi=200)  
for entity in entities:
    index_start = 0
    ix[entity] = []
    iy[entity] = []
    sum = 0
    for index in range(0,len(values[entity])):
        now   = dt_list[entity][index]
        count = 1
        mysum = 0
        breaker = 0
        while True:
            breaker += 1
            if breaker >= 1000:
                break
            try:
                mysum += values[entity][index-count-1]
                count += 1
                timedelta = now-dt_list[entity][index-count-1]
                if timedelta.seconds/(N*60) >= 1:
                    break
            except IndexError:
                break
        ix[entity].append(timestamps[entity][index])
        iy[entity].append(mysum/(count-1))

         
    plt.plot_date(x=ix[entity], y=iy[entity], fmt="-", label=entity)
    
plt.grid(b = 'on')
plt.style.use('fivethirtyeight')
plt.legend(loc='lower left')
plt.ylabel('Rolling Average of '+data_type)
plt.xlabel('Time')
```
